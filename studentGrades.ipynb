```python
import pandas as pd

#-- Read grades from file
grades = pd.read_csv('grades.csv')
grades = grades.set_index(['Course', 'StuID'])
#print(grades)

#-- Read student information from file
stuInfo = pd.read_csv('stuInfo.csv')
#print(stuInfo)
```

## Students by Course ##

```python
grades = grades.sort_index()
grades
```

## Average scores for each course ##

```python
grades.mean(level='Course')
```

## Grades including student names ##

```python
mergedDF = grades.reset_index().merge(stuInfo.reset_index(),
                   left_on=['StuID'],
                   right_on=['id']
                  ).set_index(['Course','StuID']).drop(columns=['Unnamed: 0','index','id'])
mergedDF.sort_values('major')
```

## Data Science Students ##

```python
mergedDF[mergedDF['major'] == 'Data Science']
```

## Aggregation Tasks ##

```python
# -- function to display top n records, based on a specific column
# - NOTE: default is to sort/display
def top(df, n=5, column='Final'):
    return df.sort_values(by=column)[-n:]

# -- Grouped by majors
majorGroup = mergedDF.groupby('major')
majorGroup.apply(top)
```

```python
# -- Count for each major
majorGroup.size()
```

```python
# -- Average Final Exam for each major
majorGroup['Final'].mean()
```

```python
#-- Display course rosters with grades
from IPython.display import display
courseGroup = mergedDF.groupby('Course')
for name, group in courseGroup:
    print('Course Name: ', name)
    print('Student Count: ', len(group))
    print('Final Exam (mean): ', group['Final'].mean())
    display(group)
    print('--------------------------------------------------------------------------------\n')
```
